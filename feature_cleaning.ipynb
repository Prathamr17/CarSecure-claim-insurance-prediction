```python
def length(l) :
    return int(l[l.index(')')+1 :]) / 1000
```

```python
def width(w):
    return int(w[: w.index('(')]) / 1000
```

```python
df['Length'] = df['Length'].apply(length)
df['Length']
```

```python
df['width'] = df['width'].apply(width)
```

```python
df['width']
```

```python
df['height'] = df['height'] / 1000
df['height']
```

```python
def weight(w):
    return int(w[:w.index('K')]) / 1000
    
```

```python
df['Gross_weight'] = df['Gross_weight'].apply(weight)
```

```python
def torque_rpm_ratio(s):
    torque = float(s[:s.index('N')])
    rpm = float(s[s.index('@')+1 : s.index('r')])
    return torque / rpm
```

```python
df['max_torque'] = df['max_torque'].apply(torque_rpm_ratio)
```

```python
def bhp_rpm_ratio(s):
    torque = float(s[:s.index('b')])
    rpm = float(s[s.index('@')+1 : s.index('r')])
    return torque / rpm
```

```python
df['max_power'] = df['max_power'].apply(bhp_rpm_ratio)
```

```python
df['population_density'] = (
    df['population_density']
      .astype(str)
      .str.replace('"', '')
      .str.replace('_', '')
      .str.strip()
      .replace('', None)
      .astype(int)
)
```

```python
df = df.drop('policy_id', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
df['area_cluster'] = LabelEncoder().fit_transform(df['area_cluster'])
```
